```python
#Solution2
import matplotlib.pyplot as plt
data = {'Year': [1920,1930,1940,1950,1960,1970,1980,1990,2000,2010],
        'Unemployment_Rate': [9.8,12,8,7.2,6.9,7,6.5,6.2,5.5,6.3]
       }
xPoints = data['Year']
yPoints = data['Unemployment_Rate']
plt.bar(xPoints, yPoints, label="Year", color='b', width=3)
plt.plot(xPoints, yPoints, label="Rate", color='g')
plt.title("Unemployment Rate")
plt.xlabel('Year')
plt.ylabel('Unemployment_Rate')
plt.plot()
#plt.legend()
```

```python
#Solution3, side by side using matplotlib
import matplotlib.pyplot as plt
data = {'Year': [1920,1930,1940,1950,1960,1970,1980,1990,2000,2010],
        'Unemployment_Rate': [9.8,12,8,7.2,6.9,7,6.5,6.2,5.5,6.3]
       }
xPoints = data['Year']
yPoints = data['Unemployment_Rate']
plt.tight_layout()
f, (ax1, ax2) = plt.subplots(1, 2, sharey=False, figsize=(16, 5))
ax1.plot(xPoints, yPoints,  color='g')
ax1.set_title('Unemployment Rate Matplotlib Line Chart')
ax1.set_xlabel('Year')
ax1.set_ylabel('Unemployment_Rate')
ax2.bar(xPoints, yPoints, color='b', width=4)
ax2.set_title('Unemployment Rate Matplotlib Bar Chart')
ax2.set_xlabel('Year')
ax2.set_ylabel('Unemployment_Rate')
```

```python
#Solution4 using pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt
data = {'Year': [1920,1930,1940,1950,1960,1970,1980,1990,2000,2010],'Unemployment_Rate': [9.8,12,8,7.2,6.9,7,6.5,6.2,5.5,6.3]}
df = pd.DataFrame(data,columns=['Year','Unemployment_Rate'])

fig2, ax2 = plt.subplots(nrows=1, ncols=2)
fig2.set_size_inches(20, 6)
plt.subplots_adjust(wspace=0.2)

df.plot(kind='bar', x='Year', y='Unemployment_Rate', ax=ax2[0], title='Unemployment Rate Pandas Bar Chart',color='blue')
ax2[0].set_xlabel('Year')
ax2[0].set_ylabel('Unemployment_Rate')

df.plot(kind='line', x='Year', y='Unemployment_Rate', ax=ax2[1], title='Unemployment Rate Pandas Line Chart',color='green')
ax2[1].set_xlabel('Year')
ax2[1].set_ylabel('Unemployment_Rate')

```
